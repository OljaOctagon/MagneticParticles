```python
column_name = np.round(X[all_unique_peaks]*180/np.pi,2)



df_peaks = pd.DataFrame(Y[:,all_unique_peaks], columns=column_name)
df_peaks['file_id'] = df.file_id.values
```

```python
df_merged_peaks = df_all.merge(df_peaks, on="file_id", how="inner")
df_merged_peaks = df_merged_peaks.fillna(0)
```

```python
from sklearn.manifold  import TSNE 
from sklearn.utils import shuffle 

Xpeaks = df_merged_peaks.iloc[:,3:].values
labels = df_merged_peaks["lambda"] + df_all["shift"].values


Xpeaks, labels = shuffle(Xpeaks,labels,random_state=0)

pembedding = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=12, n_jobs=8).fit_transform(Xpeaks)
import matplotlib.pyplot as plt 

color_dict = { 100.4:'red', 10.4: 'blue', 100 : 'magenta', 10 :'green'}
plt.scatter( pembedding[:,0], pembedding[:,1], color=[ color_dict[i] for i in labels ], s=1)
plt.legend(loc="lower left")
plt.show()

fig,ax = plt.subplots()
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(Xpeaks)
color_dict = { 100.4:'red', 10.4: 'blue', 100 : 'magenta', 10 :'green'}
plt.scatter( X_pca[:,0], X_pca[:,1], color=[ color_dict[i] for i in labels ], s=1)
plt.legend(loc="lower left")
plt.show()
```

```python
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
print("Loadings:")
print(loadings)

import seaborn as sns
import matplotlib.pyplot as plt

feature_names = df_merged_peaks.iloc[:,3:].columns


plt.figure(figsize=(10, 8))
sns.heatmap(loadings, annot=True, cmap='coolwarm', xticklabels=['PC1', 'PC2'], yticklabels=feature_names)
plt.title('Feature Importance in Principal Components')
plt.show()
```

```python
def biplot(score, coeff, labels, features):
    xs = score[:, 0]
    ys = score[:, 1]
    plt.figure(figsize=(10, 8))
    color_dict = { 100.4:'red', 10.4: 'blue', 100 : 'magenta', 10 :'green'}
    plt.scatter( xs,ys, color=[ color_dict[i] for i in labels ], s=5)

    for i in range(len(coeff)):
        plt.arrow(0, 0, 10*coeff[i, 0], 10*coeff[i, 1], color='r', alpha=0.5)
        plt.text(coeff[i, 0] * 10, coeff[i, 1] * 10, features[i], color='k',fontsize=6)

    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title("Biplot")
    plt.xlim((-10,10))
    plt.ylim((-10,10))

    plt.show()


from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(Xpeaks)
biplot(X_pca, pca.components_.T, labels, feature_names)
```
