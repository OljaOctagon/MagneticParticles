```python
# plot all orientation histograms 

import matplotlib.pyplot as plt

# ---- Drop the last three columns ----
df_trim = df.iloc[:, :-3]          # or:  df.drop(df.columns[-3:], axis=1)

# ---- 32 curves: each original row is a curve ----
df_trim.T.plot(legend=False)       # transpose  rows become columns
plt.xlabel("column (xaxis)")
plt.ylabel("value")
plt.show()
```

```python
Y = df_trim.values.astype(float)
X = df_trim.columns.values.astype(float)
X = np.insert(X,0,-0.05)
label = df["lambda"] + df["shift"]

color_dict = { 100.4:'red', 10.4: 'blue', 100 : 'magenta', 10 :'green'}
```

```python
from scipy.signal import find_peaks
all_peaks=[]
for y,li in zip(Y,label):
    y = np.insert(y,0,0)
    peaks, heights = find_peaks(y, height=0.05)
    
    all_peaks.extend(peaks)
    print(peaks)
    print(heights)
    print(X[peaks]*180/np.pi)
    #fig,ax = plt.subplots()
    plt.plot(X,y,c=color_dict[li])
    plt.plot(X[peaks],y[peaks], "x")

 
plt.show()
    

all_peaks = np.array(all_peaks)
all_unique_peaks = np.unique(all_peaks)

print(X[all_unique_peaks]*180/np.pi)
print(Y[0][all_unique_peaks])
    
    
    
```

```python
dir="/Users/ada/Documents/github_repos/magnetic_particles/MagneticParticles/rigid_magnetic/example_data"
filen="MAG2P_nbonded-2025-3-13-15:3:59.pickle"

df_all = pd.read_pickle("{}/{}".format(dir,filen)) 
```

```python
df_all
```

```python
df_merged = df_all.merge(df, on=["file_id", "lambda","shift"], how="inner")
df_merged = df_merged.iloc[:,:-24]
```

```python
df_merged = df_merged.fillna(0)
```

```python
df_merged
```

```python
from sklearn.manifold  import TSNE 
from sklearn.utils import shuffle 

X = df_merged.iloc[:,3:].values
df_merged["label"] = df_merged["lambda"] + df_merged["shift"]
labels = df_merged["label"].values

X, labels = shuffle(X,labels,random_state=0)

embedding = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=10, n_jobs=8).fit_transform(X)
```

```python
labels
```

```python
import matplotlib.pyplot as plt 
from sklearn.manifold import TSNE
color_dict = { 100.4:'red', 10.4: 'blue', 100 : 'magenta', 10 :'green'}
plt.scatter( embedding[:,0], embedding[:,1], color=[ color_dict[i] for i in labels ], s=1)
plt.legend(loc="lower left")
#plt.xlim((-50,50))
#plt.ylim((-50,50))
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
color_dict = { 100.4:'red', 10.4: 'blue', 100 : 'magenta', 10 :'green'}
plt.scatter( X_pca[:,0], X_pca[:,1], color=[ color_dict[i] for i in labels ], s=1)
plt.legend(loc="lower left")
plt.show()

```
